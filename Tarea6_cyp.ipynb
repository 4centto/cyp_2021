```
#Problema 3.3

N = int(input("Dame un numero: "))
FLAG = True
i = 1
ACUM = 0

while i <= N:
    if FLAG == True:
        ACUM += (1 / i)
        FLAG = False
        i += 1

    else:
        ACUM -= (1 / i)
        FLAG = True
        i += 1

print("El resultado de la sucesion es: ", ACUM)
print("Programa finalizado")
```

```
#Problema 3.13
import random

#Inicializacion de los arreglos
Norte = [0.0 for i in range(12)]
Centro = [0.0 for i in range(12)]
Sur = [0.0 for i in range(12)]

##Llenado de los arreglos
for i in range(12):
    Norte[i] = random.randint(100, 200)
    Centro[i] = random.randint(100, 200)
    Sur[i] = random.randint(100, 200)

#Impresion para la visualizacion de los datos
print("Norte: ", Norte)
print("Centro: ", Centro)
print("Sur: ", Sur)
print()

#Promedio anual de la region centro
AnualCentro, PROMEDIOANUALCENTRO = 0, 0

for i in range(len(Centro)):
    AnualCentro += Centro[i]
PROMEDIOANUALCENTRO = AnualCentro / 12
print("El promedio anual del centro es de: ", round(PROMEDIOANUALCENTRO, 2))

#Mes y registro con menor lluvia en la region sur
MES, REGISTRO = 0, 1

for i in range(len(Sur)):
    if Sur[i] > REGISTRO:
        REGISTRO = Sur[i]
        MES = (i + 1)

print("En el mes ", MES, " se registro mayor promedio en el sur con: ", REGISTRO)

#Region con mayor lluvia anual
norte, centro, sur = 0, 0, 0

for i in range(12):
    norte += Norte[i]
    centro += Centro[i]
    sur += Sur[i]

if norte > centro and norte > sur:
    print("El norte es la region con mayor promedio de lluvias con: ", norte)
elif centro > sur:
    print("El centro es la region con mayor promedio de lluvias con: ", centro)
else:
    print("El sur es la region con mayor promedio de lluvias con: ", sur)
```
